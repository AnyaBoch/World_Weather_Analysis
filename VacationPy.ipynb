## Starter Code to Import Libraries and Load the Weather and Coordinates Data

```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
city_data_df = pd.read_csv("output_data/cities.csv")

# Display sample data
city_data_df.head()
```

---

### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
# %%capture --no-display

# Configure the map plot
city_map_plot = city_data_df.hvplot.points(
    "Lng", "Lat",  # x and y axis
    geo=True,  # enable geospatial plotting
    tiles="OSM",  # use OpenStreetMap tiles
    size="Humidity",  # size of points based on Humidity
    color="City",  # optional for visual distinction by city
    cmap="set3",
    frame_width=800, 
    frame_height=600,
    title="City Map (Humidity)"
)

# Display the map
city_map_plot
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values
ideal_conditions = city_data_df[
    (city_data_df["Max Temp"] >= 10) & # min temp of my "ideal" filter
    (city_data_df["Max Temp"] <= 28) & # max temp of filter
    (city_data_df["Wind Speed"] < 7)& # wind speed should be less than 7 ((unfortunately we do not have units,))
     (city_data_df["Humidity"] > 60) & # moderate humidity
    (city_data_df["Cloudiness"]<=40)
]
# Drop any rows with null values
ideal_conditions = ideal_conditions.dropna()

# Display sample data
ideal_conditions
```

```python
#let's do geomap plot for my ideal weather conditions just for fun 
# with the same Humidity value for city size

# %%capture --no-display

# Configure the map plot for ideal conditions
ideal_map_plot = ideal_conditions.hvplot.points(
    "Lng", "Lat",  # x and y axis
    geo=True,  # enable geospatial plotting
    tiles="OSM", 
    size="Humidity",  
    color="City",  
    cmap="set3",  
    frame_width=800, 
    frame_height=600,
    title="Ideal Weather Conditions Map"
)

# Display the map
ideal_map_plot
```

```python
#let's count the number of cities that fits conditions (actually should be as many as rows DF "ideal_conditions"  )
number_of_ideal_cities = ideal_conditions["City"].count()
print(f"Number of cities with ideal conditions: {number_of_ideal_cities}")
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = ideal_conditions[["City", "Country", "Lat", "Lng", "Humidity"]].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df["Hotel Name"] = ""

# Display sample data
hotel_df.head()
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000 #as we are already in metric units 
limit = 1
categories = "accommodation.hotel"

params = {
        "categories": categories,
        "limit": limit,  # Retrieving only the first result
        "apiKey": geoapify_key
    }


# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame

for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    lat = hotel_df.loc[index,"Lat"]
    lng = hotel_df.loc[index,"Lng"]

    # Add the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{lng},{lat},{radius}"
    params["bias"] = f"proximity:{lng},{lat}"

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make and API request using the params dictionary
    name_address = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    name_address = name_address.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```
